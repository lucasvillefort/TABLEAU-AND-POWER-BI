```python
import pandas as pd
import numpy as np
```

```python
database  = pd.read_csv('PART1_EXCEL.xlsx')    
database
```

```python
database.dtypes
```

```python
for column in database.columns:
    if '\t' in column:
        database = database.rename(columns={column: column.replace('\t', ' ')}) 
       
    if database[column].dtype == 'object':
        database[column] = database[column].replace(r'\t', ' ', regex=True)
    if database[column].dtype == 'datetime64[ns]':
        database[column] = database[column].dt.strftime('%Y-%m-%d') 

# assuming 'df' is your DataFrame
```

```python


# write the modified DataFrame back to the original file
database.to_csv("PART1_EXCEL.csv", index=False)
```
